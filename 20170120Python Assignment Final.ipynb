## 1. Research Question

In their paper [*“Immigration, Offshoring and American Jobs”*](https://www.aeaweb.org/articles?id=10.1257/aer.103.5.1925), published in 2013, Gainmarco Ottaviano, Giovanni Peri and Greg Wright analyze the impact of immigration and offshoring on employment in the American manufacturing sector. They find that immigration and offshoring have a positive effect on total emplyoment in the US manufacturing sector. 
In the following Notebook, first, their theoretical model is described, then some descriptive statistics are provided and finally some of their empirical results are replicated and presented. In particular, it is evaluated what effect an decrease in the costs of immigrant and offshoring labor have on the shares of domestic, immigrant and offshore workers as well as on the total employment level.


## 2. Motivation 

It  is  a  widespread  believe  that  increases  in  offshoring  and  immigration  are  credited  with  a  demise  of  manufacturing  jobs  of  American  natives:  asked  whether  the  developments  of  offshoring  and  immigration  have  done  more  to  help  or  to  hurt  American  workers,  77%  of  the  Americans  claim  that  offshoring  has  hurt  them  and  55%  say  that  immigration  has  hurt  them ([PewReseachCenter 2006](http://www.pewsocialtrends.org/2006/08/30/american-work-life-is-worsening-but-most-workers-still-content/)). The newly elected American president, Donald Trump also has this vision on Migration: he wants to put America first, meaning that American citizens will get the jobs first. ([Trump pence, 2016](https://www.donaldjtrump.com/policies/immigration)) Thus, it is an interesting and important topic to investigate. Moreover, there are several recent papers focusing either on the effects of immigration or on the effects of offshoring on the American labor market. However, there is no paper providing a combined empirical analysis of these two aspects of globalisation on employment. [Ottaviano et al.](https://www.aeaweb.org/articles?id=10.1257/aer.103.5.1925) claim that "[...] the joint investigation of the interactions among these three groups of workers is useful in order to improve our understanding of the impact of globalization on the US labor market and, in particular to answer to hotly debated questions. First, how do declines in offshoring and immigration costs affect the employment of native workers? Second, what kinds of jobs suffer or benefit the most form the competition created by offshore and immigrant wokers?" In the following, it is focused on the results respective to the first of these questions.

Based on this interpretation, the assumptions of the model are  specified.

** Table 1. Exogenous parameters** 

|Symbol| Definition| 
|------|------|------|
|$w$ | Wage native workers|
|$\tilde{w}$| Wage immigrant worker| 
|$w^*$| Wage offshore worker|
|$a_L$| Unit labor requirement|
|$\delta$| Foregone productivity M|
|$\tau(i)$| Unit input requirement M|
|$\beta t(i)$| Unit input requirement O|


** Table 2. Endogenous parameters**

|Symbol| Definition|
|------|------|------|
|$c_D(i)$| Marginal costs producing task $i$ employing native workers|
|$c_M(i)$|  Marginal costs producing task $i$ employing migrant workers|
|$c_O(i)$|  Marginal costs producing task $i$ employing offshore workers|



<br>
 
Ottaviano  et  al.   developed  a  partial  equilibrium  model. This  theoretical  model  describes  the  distribution  of  jobs  among  native,  immigrant  and offshore  employees  depending  on  the  complexity  of  the  task,  while  only  manufacturing  tasks, requiring low skills, are considered.  
Natives,  immigrants  and  offshore  workers  are  considered  to  be  perfectly  substitutable.  Thus, the worker entailing the lowest marginal cost will  perform a task in the equilibrium. The marginal  costs for natives are determined only by their wage $w$, while the marignal production costs of immigrant and offshore workers are composed by their wages $\tilde{w}$, $w^*$, frictional costs $\delta$, $\beta$ and task specific costs $\tau(i)$, $t(i)$. 

For all three groups of workers the labor unit requirment $a_L$ is the same. This results in the following marginal production costs:

$c_D=wa_L$, 

$c_M=\tilde{w}\tau(i)a_L=\delta w^*\tau(i)a_L$ 

$c_O=w^*\beta t(i)a_L$. 

Moreover, Ottaviano et al. specified sufficient conditions on the characteristics of the components of the marginal costs for the three groups of workers to specialize in high, low and medium complexity tasks. Thus, a marginal immigrant task $I_{MO}$ as well as a marginal offshore task $I_{NO}$ exists. 
<br>
Hence, the marginal costs for the three types of workers depending on the complexity of the tasks and the resulting assignment of the tasks to natives, immigrants and offshore workers can be illustrated like this:


**Figure 4.**
![Task distribution according to task complexity](./helene.png)

From this figure it is clear that $I_{MO}$ and $I_{NO}$ can be specified as follows:

$I_{MO}$ is reached when $c_M(i)$ = $c_O(i)$, so if

$\delta w^*\tau(i)a_L$ = $w^*\beta t(i)a_L$

$I_{NO}$ is reached when $c_D$ = $c_O(i)$, so if

$wa_L$ = $w^*\beta t(i)a_L$

### 3.2 Predictions of the Model

Based on this model, predictions can be made about the effects of changes in the sector specific cost parameters $\delta$ an $\beta$, which are referred to as changes in the costs of immigration and offshoring, on employment shares, employment levels and task assignment. As mentioned before, the focus here will be on the effects on employment shares and employment levels. 
Lower immigration costs are expected to reduce the share of offshore workers, to have no effect on the share of natives and to increase the share of immigrants. Lower offshoring costs are expected to reduce the shares of natives and immigrants, and to increase the share of offshore workers. Moreover, as a consequence of declining offshoring or immigration costs, the demand for workers is expected to rise, so that an increase in the total sector employment is expected. 

## 4. The Econometric Model 

### 4.1 Displacement effect 

First,  in  order  to  explore  the  degree  of  direct  competition  between  the  three  groups  of  workers,  also  referred  to  as  the  “displacement  effect”,  the  effect  of  a  change  in  immigration  and  offshoring  costs  on  the  different  employment  shares  is  estimated.  Therefore, the following regressions are estimated:  
<br>

Equation (1)
$$
{s_D}_{st}=\phi_{s}^{D}+\phi_{t}^{D}+b_{DO}({s_O}_{st})+b_{DM}({s_M}_{st})+\epsilon_{st}^{D}
$$
<br>
Equation (2)
$$
{s_M}_{st}=\phi_{s}^{M}+\phi_{t}^{M}+b_{MO}({s_O}_{st})+\epsilon_{st}^{M}
$$
<br>
Equation (3)
$$
{s_O}_{st}=\phi_{s}^{O}+\phi_{t}^{O}+b_{OM}({s_M}_{st})+\epsilon_{st}^{O}
$$
<br>


Here ${s_D}_{st}$, ${s_M}_{st}$ and ${s_O}_{st}$ are the shares of domestic, immigrant and offshore workers by industry $s$ and year $t$. $\phi_{s}$ are industry fixed effects and $\phi_{t}$ are time fixed effects. $\epsilon_{st}$ are potentially serially correlated standard errors. 




### 4.2 Productivity effect

Second,  the  effect  of  changes  in  immigration  and  offshoring  costs  on  the  total  employment  level,  also  referred  to  as  the  “productivity  effect”,  is  estimated  with  the  following regression: 
<br><br>


Equation (4) $$
{N_L}_{st}=\phi_{s}^{L}+\phi_{t}^{L}+B_{LO}({s_O}_{st})+B_{LM}({s_M}_{st})+\epsilon_{st}^L  
$$

<br>

Here ${N_L}_{st}$ is the logarithm of the total employment level. 

The  productivity  effect  claims  that  because  of  lower  costs  for  hiring  immigrants  or  offshore  workers  total  costs  decline,  which  improves  the  efficiency  of  production  and  therefore induces an increased demand for all worker types. 


## 5. Reproduction of the Results 

### 5.1 Effects of Immigration and Offshoring on Employment Shares 

First,  the  effects  of  changes  in  immigration  and  offshoring  costs  on  the  employment 
shares of the three different groups of workers are estimated. This is done to explore the  relative  substitutability  of  the  three  different  groups  of  workers,  also  called  the  “displacement effect”. Empirically this is implemented by estimating the regressions (1) to (3). It has to be noted that because of the collinearity of the shares of immigrants and  offshore  workers,  equation  (1)  cannot  be  estimated  in  the  form  described before.  Two separate equations, one describing the effect of the immigrant share on natives, and  one  describing  the  effect  of  the  offshore  share  on  natives,  are  estimated.  Equations  (1);  (3)  are  estimated  via  2SLS  using  the  instruments ${(Imputed_{s_M}})_{st}$
for  the  share  of  immigrants and $(Tariffs)_{st}$ for the share of offshore workers. The resulting coefficients  are  thus  easily  interpretable:  They  show  the  percentage  change  in  native  (immigrant/offshore)  employment  resulting  out  of  a  1-percentage  change  in  immigrant/offshore  employment.  In  all  regressions  industry  and  time  fixed  effects  are  included.  For  all  regressions  the  calculated  standard  errors  are  heteroscedasticity  robust.

Before the regressions of interest can be run, the data must be prepared. Therefore, first, the dataset that contains the employment data for natives and immigrants is used. It needs to be merged with the dataset containing the data on the instrument for the share of immigrants ${(Imputed_{s_M}})_{st}$, and with a dataset with information about the employment of offshore workers. All these datasets contain the same year and industry variables, thus they can be merged by them. Only employment in foreign affiliates of US parent companies is considered, not included is employment by foreign subcontractors (also called arm´s length offshoring). Therefore, the variables emp armemp and totemp get dropped. This dataset gets merged with a dataset containing the instrument for the share of offshore employment. After that the shares of employment get constructed, while for natives and immigrant workers only workers with high school diploma or lower education are considered. This dataset can be used for the regressions of the "productivity effect" as well.

**Include Regressions**

```python
import numpy as np
import pandas as pd 
import statsmodels.formula.api as sm #import of the required packages 

data = pd.read_csv("/Users/Sietske/Documents/data00_07.csv") #import the data 

result= sm.ols(formula="share_nat ~ share_off", data=data).fit() #running the OLS 

print(result.params)

print(result.summary())
```
